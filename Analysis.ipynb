```python
job_skills_data = job_skills.groupby('job_id')['skill_abr'].agg(lambda x: ', '.join(x)).reset_index() # aggregate same job skills

job_postings_data = job_postings.merge(job_skills_data, on="job_id", how="left")
job_postings_data
```

```python
#company_industries_data = pd.read_csv("/kaggle/input/linkedin-job-postings/company_details/company_industries.csv")
company_industries_data = company_industries.groupby('company_id')['industry'].agg(lambda x: ', '.join(x)).reset_index() # aggregate same company industries

#company_specialities_data = pd.read_csv("/kaggle/input/linkedin-job-postings/company_details/company_specialities.csv")
company_specialities_data = company_specialities.groupby('company_id')['speciality'].agg(lambda x: ', '.join(x)).reset_index() # aggregate same company specialities

#employee_counts_data = pd.read_csv("/kaggle/input/linkedin-job-postings/company_details/employee_counts.csv")
employee_counts_data = employee_counts.groupby('company_id')['time_recorded'].max().reset_index() # get newest data based on 'time_recorded' column

company_data = companies.merge(company_industries_data, on="company_id", how="left")
company_data = companies.merge(company_specialities_data, on="company_id", how="left")
company_data = companies.merge(employee_counts_data, on="company_id", how="left")
company_data
```

```python
merged_data = job_postings_data.merge(company_data, on="company_id", how="left")
merged_data
```

```python
merged_data.isnull().sum()
```

```python
data = merged_data.dropna(subset=['company_id'])
data.info()
```

```python
# Handling missing values

# Use "Not Specified" for categorical columns with missing values
cols_fill_not_specified = ['skills_desc', 'pay_period', 'currency', 'compensation_type', 'posting_domain',
                           'application_url', 'formatted_experience_level', 'company_size', 'zip_code', 'address',
                           'state', 'url', 'city', 'country', 'name']
for col in cols_fill_not_specified:
    data[col].fillna("Not Specified", inplace=True)

# Fill numerical columns with zeros
cols_fill_zero = ['applies', 'views']
for col in cols_fill_zero:
    data[col].fillna(0, inplace=True)

# Fill remote_allowed with "Unknown"
data['remote_allowed'].fillna("Unknown", inplace=True)
```
